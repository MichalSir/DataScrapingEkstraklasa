```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from datetime import datetime
```

```python
month = {"maj": "5", "kwi": "4", "mar": "3", "lut":"2", "sty": "1",
         "gru": "12", "lis": "11", "paź": "10", "wrz": "9", "sie": "8"}
urlBase = "https://www.meczyki.pl"
urlSchedule = 'https://www.meczyki.pl/ekstraklasa,119,terminarz'
urlTeam = "https://www.meczyki.pl/wyniki-na-zywo/sklady-meczowe/"
events ={
    "c-ico c-ico-goal c-ico-16": "Gol",
    "c-ico c-ico-goal c-ico-16 cancelled": "Anulowany gol",
    "c-ico c-ico-yc c-ico-16": "Zolta kartka",
    "c-ico c-ico-y2c c-ico-16": "Druga zolta",
    "c-ico c-ico-rc c-ico-16": "Czerwona",
    "c-ico c-ico-pm c-ico-16": "Niestrzelony karny",
    "c-ico c-ico-siso c-ico-16": "Zmiana"    
}
```

```python
def CutLink(x):
    link = x.find("a")['href']
    linkStats = urlBase + link
    link = link.split("/")
    linkTeam = urlTeam + link[2] + '/' + link[3]
    return linkStats, linkTeam
```

```python
def FixTime(time):
    time = time.lstrip('\n').strip().partition('\n')[0]
    time = time.split()
    time[1] = month.get(time[1])
    time = "".join(time)
    date_time = datetime.strptime(time, '%d%m%Y|%H:%M')
    return date_time  
```

```python
def Stats(link, matchID, roundNr):
    temp_source = requests.get(link).text
    temp_soup = BeautifulSoup(temp_source, "lxml")
    time = temp_soup.find('div',attrs={'class':"section"}).find_all('div')[1].text
    date = FixTime(time)
    score = temp_soup.find('div',attrs={'class':"title u-center"}).text.split()
    score1 = score[0]
    score2 = score[2]
    if score1 != '?':
        score1 = int(score1)
        score2 = int(score2)
    name = temp_soup.find_all('div',attrs={'class':"team-name"})
    stats = temp_soup.find('div',attrs={'class':"stats-body"})
    stats_rows = temp_soup.find_all('div',attrs={'class':"stats-row"})
    info = {"ID": matchID,"Kolejka":roundNr, "Data": date, "Zespol1-Nazwa":name[0].text,"Zespol1-Bramki":score1,
            "Zespol2-Nazwa":name[1].text, "Zespol2-Bramki":score2}
    team1 = {}
    team2 = {}
    for stat in stats_rows:
        temp = stat.find_all('div')
        if len(temp) == 0:
            continue
        statName = temp[1].text
        t1 = temp[0].text
        t2 = temp[2].text
        if statName == "Posiadanie piłki":
            t1 = t1.rstrip("%")
            t2 = t2.rstrip("%")
        team1["Zespol1-"+statName] = int(t1)
        team2["Zespol2-"+statName] = int(t2)
    team1.update(team2)
    info.update(team1)
    return info
```

```python
def AllEvents(url, matchID):
    matchSummary = pd.DataFrame(columns=['ID','Czas','Kto','Co','Info'])
    temp_source = requests.get(url).text
    temp_soup = BeautifulSoup(temp_source, "lxml")
    summary = temp_soup.find('div',attrs={'class':"collapse-container clearfix"})
    events =summary.find_all("div",attrs={'class':"match-event-row with-dot"})
    for e in events:
        event = ReadEvents(e)
        matchSummary = matchSummary.append(event,ignore_index=True)
    matchSummary['ID'] = matchID
    return matchSummary
```

```python
def ReadEvents(event):
    matchSummary = pd.DataFrame(columns=['Czas','Kto','Co','Info'])
    smallEvents = event.find_all("div",attrs={'class':"match-event-box-container"}) 
    team1 = smallEvents[0].find_all("div",attrs={'class':"event-box-row"})
    team2 = smallEvents[1].find_all("div",attrs={'class':"event-box-row"})
    wpisy = list()
    if(len(team1) > 0 ):
        time = team1[0].find("div",attrs={'class':"minute"}).text.rstrip("'")
        for smallEvent in team1:
            singleInfo = Event(smallEvent, time)  
            matchSummary = matchSummary.append(singleInfo,ignore_index=True)
            
    if(len(team2) > 0 ):
        time = team2[0].find("div",attrs={'class':"minute"}).text.rstrip("'")
        for smallEvent in team2:
            singleInfo = Event(smallEvent, time)
            matchSummary = matchSummary.append(singleInfo,ignore_index=True)
    return matchSummary
```

```python
def Event(event, time):
    time = int(time)
    info = "brak"
    temp = " ".join(event.find("i")['class'])
    what = events.get(temp)
    temp2 = event.find_all("div",attrs={'class':"name-line"})
    if str(temp2[0].find("a")) != "None":
        who = temp2[0].find("a")["href"]
    else:
        who = temp2[0].text
    if what == "Zmiana" and len(temp2) == 2:
        if str(temp2[1].find("a")) != 'None':
            info = temp2[1].find("a")["href"]
        else:
            info = temp2[1].text.lstrip("(").rstrip(")") 
    if what == "Gol" and len(temp2) == 2:
        info = temp2[1].text.lstrip("(").rstrip(")") 
    singleInfo = {"Czas": time, "Kto": who, "Co": what, "Info":info}
    return singleInfo
```

```python
def GetTeamsLineup(link, matchID):
    temp_source = requests.get(link).text
    temp_soup = BeautifulSoup(temp_source, "lxml")
    players  = temp_soup.find_all("td",attrs={'class':"list-team cell-bottom-border"})
    teams = pd.DataFrame(columns=['matchID','Zespol','Sklad','Zawodnik'])
    i = 0
    team = ""
    playerID = ""
    lineup = "Podstawowy"
    for player in players:
        if i == 22:
            lineup = "Rezerwowy"
        if i % 2 == 0:
            team = "Gospodarz"
        else:
            team = "Goscie"
        
        if player.find("a") == None:
            playerID = player.text.strip()
        else:
            playerID = player.find("a")["href"]
        i += 1
        singleInfo = {"Zespol": team, "Sklad": lineup, "Zawodnik":playerID}
        teams = teams.append(singleInfo, ignore_index=True)
    teams['matchID'] = matchID
    return teams 
```

```python
#ALL MATCHES IN SCHEDULE
schedule = requests.get(urlSchedule).text
soupSchedule = BeautifulSoup(schedule, "lxml")
container  = soupSchedule.find("tbody",class_ = "round-container")
tabel = container.find_all("td",class_="status" )
```

```python
statsColumns = ['ID', 'Kolejka', 'Data', 'Zespol1-Nazwa', 'Zespol1-Bramki', 'Zespol2-Nazwa', 'Zespol2-Bramki',
           'Zespol1-Posiadanie piłki', 'Zespol1-Strzały', 'Zespol1-Strzały celne', 'Zespol1-Strzały niecelne',
           'Zespol1-Strzały zablokowane', 'Zespol1-Ataki', 'Zespol1-Groźne ataki', 'Zespol1-Kartki', 'Zespol1-Faule',
           'Zespol1-Rzuty wolne', 'Zespol1-Rzuty rożne', 'Zespol1-Spalone', 'Zespol2-Posiadanie piłki', 'Zespol2-Strzały',
           'Zespol2-Strzały celne', 'Zespol2-Strzały niecelne', 'Zespol2-Strzały zablokowane', 'Zespol2-Ataki',
           'Zespol2-Groźne ataki', 'Zespol2-Kartki', 'Zespol2-Faule', 'Zespol2-Rzuty wolne', 'Zespol2-Rzuty rożne',
           'Zespol2-Spalone']
```

```python
roundNr = 30
matchID = 0
stats = pd.DataFrame(columns=statsColumns)
matchSummary = pd.DataFrame(columns=['ID','Czas','Kto','Co','Info'])
teams = pd.DataFrame(columns=['matchID','Zespol','Sklad','Zawodnik'])
for x in tabel:
    statsLink, teamLink = CutLink(x)
    s = Stats(statsLink, matchID, roundNr)
    stats = stats.append(s,ignore_index=True)
    aE = AllEvents(statsLink, matchID)
    matchSummary = matchSummary.append(aE,ignore_index=True)
    tL = GetTeamsLineup(teamLink, matchID)
    teams = teams.append(tL, ignore_index=True)
    matchID += 1
    
    if matchID % 8 == 0:
        roundNr -= 1
```

```python
stats.to_csv(r'StatsTemp.csv', index = False,sep=';')
matchSummary.to_csv(r'SummaryTemp.csv', index = False,sep=';')
teams.to_csv(r'TeamsTemp.csv', index = False,sep=';')
```
